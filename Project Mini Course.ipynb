```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Data and Cleaning Data

```python
#how to read data in google sheet
sheet_url = 'https://docs.google.com/spreadsheets/d/12gwnJPDKBIzZ2nD4NOr0e5n-GqNPRn9GA25sB6FvJQg/edit#gid=1903509374'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
print(sheet_url_trf)

df = pd.read_csv(sheet_url_trf)
df.head()
```

```python
#change typo date
df_cleaned = df.copy()

df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])
df_cleaned['Revenue'] = df_cleaned['Unit price'] * df_cleaned['Quantity']

print(df_cleaned.info())
df_cleaned.head()
```

```python
#check typo
for col_name in ['Payment', 'Branch', 'City', 'Customer type', 'Gender', 'Product line']:
  print (df_cleaned[col_name].value_counts(), '\n')
```

```python
#replace
dict_typo = {'male' : 'Male',
             'female' : 'Female',
             'fml' : 'Female'}

df_replaced = df_cleaned.replace(dict_typo)
df_replaced.head()
```

```python
df_replaced['Gender'].value_counts()
```

```python
df_replaced['Invoice ID'].duplicated().sum()
```

# EDA

### daily trend overall

```python
agg_daily = df_replaced.groupby('Date', as_index=False)['Revenue', 'Quantity'].sum()
agg_daily
```

```python
plt.rcParams["figure.figsize"] = (20, 5)

plt.plot('Date', 'Revenue', data=agg_daily)
plt.title('Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Rev($)')
plt.show()
```

```python
sns.lineplot(data=agg_daily, x='Date', y='Quantity')
```

## By Payment

```python
agg_payment = df_replaced.groupby('Payment', as_index=False).agg({'Revenue':'sum', 'Invoice ID':'nunique'})
agg_payment.head()
```

```python
plt.rcParams['figure.figsize'] = (10,5)
sns.barplot(x='Payment', y='Revenue', data = agg_payment.sort_values('Revenue', ascending=False))
```

highest revenue datang dari cash

```python
sns.barplot(x='Payment', y='Invoice ID', data = agg_payment.sort_values('Invoice ID', ascending=False))
```

berdasarkan count transaction, ewallet paling banyak

```python
plt.pie(agg_payment['Revenue'], labels = agg_payment['Payment'],autopct='%.1f%%')
```

cash valume 34.7% dan itu diatas 3.5% dari credit card
